```python
from sklearn.datasets import make_blobs
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
```

```python
x, _ = make_blobs(n_samples=5000, centers=5)
```

```python
x
```

```python
scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
```

```python
import matplotlib.pyplot as plt

plt.scatter(x_scaled[:, 0], x_scaled[:, 1])
```

```python
kmeans = KMeans(n_clusters=5)
kmeans.fit(x_scaled)
```

```python
plt.scatter(x_scaled[:, 0], x_scaled[:, 1], c=kmeans.labels_)
```

```python
from sklearn.datasets import make_moons
x, _ = make_moons(n_samples=5000, noise=0.05)

scaler = StandardScaler()
x_scaled = scaler.fit_transform(x)
```

```python
plt.scatter(x_scaled[:, 0], x_scaled[:, 1])
```

```python
plt.scatter(x_scaled[:, 0], x_scaled[:, 1], c=kmeans.labels_)
```

```python
from sklearn.cluster import DBSCAN

dbscan = DBSCAN(eps=0.03)
dbscan.fit(x_scaled)
plt.scatter(x_scaled[:, 0], x_scaled[:, 1], c=dbscan.labels_)
```
